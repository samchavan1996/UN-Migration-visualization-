```python
#Author-Sampada Chavan
#Final Project-EM624
#Imported pandas for for handling CSV files and making data frame
import pandas as pd
#Imported numpy to perform mathematical operations
import numpy as np
#Imported seaborn for various plots such as boxplot,lineplot,factorplot
import seaborn as sns
#Imported matplotlib to plot
import matplotlib.pyplot as plt

#read migration csv into df variable
df = pd.read_csv(r"UN_Migrations.csv")
print(df.head(5))
col = list(df.columns[0:6])
df.columns = df.iloc[0]
other_columns = col + list(df.columns[6:])
df.columns = other_columns
df.drop(0, inplace=True)

df.info()
#in result variable took all the datatypes of columns and printed them
result = df.dtypes
print(result)

#printing all the column names
for col in df.columns:
    print(col)
#to drop columns having all NA values
df = df.dropna(axis=1, how='all')
print(df)

#all contry columns were of object datatyoe to change datatype object to float
#all the columns are taken into num_col variable and then column values
#having .,',"",- are replaced with blank
num_col= df.columns[6:]
for column in num_col:
    df[column] = df[column].str.replace(',', "")
    df[column] = df[column].str.replace('.', "")
    df[column] = df[column].str.replace("", "")
    df[column] = df[column].str.replace("-", "")
    df[column] = df[column].replace(r'^\s*$', np.nan, regex=True)
    #here dataype is changed from object to float
    df[column] = df[column].str.strip().astype(float)

#again printing column names along with their changed datatypes
for i in df.columns[6:]:
    print(str(df[i].name) + " ---" + str(df[i].dtype))

#changing datatype of year column
df['Year'] = df['Year'].astype(str).astype(float)

#defining migration function and  passing region and year parameter
def migration(reg, y):
    #printing region and year
    print("Year:", y)
    print("Region:", reg)
    #created new dataframe named list and stored columns from Other south to zimbabwe
    #and having rows of column Major area, region, country or area of destination and year values same as given region and year
    list = df.loc[:, 'Other South':'Zimbabwe'][
        (df['Major area, region, country or area of destination'] == reg) & (df['Year'] == y)]
    # print(list)
    #creating list having 3 tuple values first value in tuple is index of row, second value is name of region and third value is the population count
    dd = list.stack().reset_index().agg(tuple, 1).tolist()
    #sorting the list having tuple by population mograted value
    dd.sort(key=lambda i: i[2], reverse=True)
    #in list dd1 taking top 5 values having highest migration value and printing it
    dd1 = dd[0:5]
    print(dd[0:5])
    #extracting the country name in country_data list and migration count in country_value list
    country_data = ([x[1] for x in dd1])
    country_value = ([x[2] for x in dd1])
    #print(country_data)
    #print(country_value)
    #defining figure size
    fig, ax = plt.subplots(figsize=(8, 4))
    #plotting barplot
    sns.barplot(x=country_data, y=country_value, ax=ax)
    #giving titlw to barplot
    plt.title("Top 5 migrating population from following countries to-->{reg}")
    #displaying the plot
    plt.show()


# following countries population are migrating to more developed regions like Europe,Northern America, Australia, New Zealand, and Japan
#running migration function for more developed regions as destination
developed_reg = ["EUROPE AND NORTHERN AMERICA", "Australia", "New Zealand", "Japan"]
years = [1990, 1995, 2000, 2005, 2010, 2015, 2019]
for i in range(len(developed_reg)):
    for j in range(len(years)):
        migration(developed_reg[i], years[j])

# # contries people migrated to less developed regions
#running migration function for less developed region as a destination such as Africa, Asia, Latin America and Caribbean,etc
undeveloped_reg = ["SUB-SAHARAN AFRICA", "NORTHERN AFRICA AND WESTERN ASIA", "CENTRAL AND SOUTHERN ASIA",
                   "EASTERN AND SOUTH-EASTERN ASIA", "LATIN AMERICA AND THE CARIBBEAN", "Melanesia", "Micronesia",
                   "Polynesia"]
years = [1990, 1995, 2000, 2005, 2010, 2015, 2019]
for i in range(len(undeveloped_reg)):
    for j in range(len(years)):
        migration(undeveloped_reg[i], years[j])

# # Based on developement
#running migration function for various types of region such as more developed,less developed
#and finding whether which country people are migrating to what type of developed region
dev_reg = ["More developed regions", "Less developed regions",
           "Less developed regions, excluding least developed countries"]
years = [1990, 1995, 2000, 2005, 2010, 2015, 2019]
for i in range(len(dev_reg)):
    for j in range(len(years)):
        migration(dev_reg[i], years[j])

# # Throughout world maximum number of people are migrating from this countries
#here we found complete trend of migration for entire world and found top 5 countries having high number of migrating population
complete = ["WORLD"]
years = [1990, 1995, 2000, 2005, 2010, 2015, 2019]
for i in range(len(complete)):
    for j in range(len(years)):
        migration(complete[i], years[j])

#
list = df.loc[(df['Major area, region, country or area of destination'] == "CENTRAL AND SOUTHERN ASIA")]
print(list)
ax = sns.lineplot(x='Total', y='Year', data=list)

#
list = df.loc[(df['Major area, region, country or area of destination'] == "SUB-SAHARAN AFRICA")]
print(list)
ax = sns.lineplot(x='Total', y='Year', data=list)

#
list = df.loc[(df['Major area, region, country or area of destination'] == "More developed regions")]
print(list)
ax = sns.lineplot(x='Total', y='Year', data=list)

#
list = df.loc[(df['Major area, region, country or area of destination'] == "Less developed regions")]
print(list)
ax = sns.lineplot(x='Total', y='Year', data=list)

#
list = df.loc[:, 'Total':'Zimbabwe'][(df['Major area, region, country or area of destination'] == "WORLD")]
print(list)
dd = list.stack().reset_index().agg(tuple, 1).tolist()
dd.sort(key=lambda i: i[2], reverse=True)
print(dd)
dd1 = dd[7:300]
# ax = sns.lineplot(x='Year', y='Total',data=dd)
country_data = ([x[1] for x in dd1])
country_value = ([x[2] for x in dd1])
fig, ax = plt.subplots(figsize=(100, 4))
sns.swarmplot(x=country_data, y=country_value, ax=ax)


#
list = df.loc[:, 'Total':'Zimbabwe'][(df['Year'] == 2019)]
dd = list.stack().reset_index().agg(tuple, 1).tolist()
dd.sort(key=lambda i: i[2], reverse=True)
print(dd[0:15])
# print(dd2[0:5])
df.loc[[1699, 1706, 1701, 1922, 1702, 1704, 1714, 1715, 1923, 1707, 1717, 1976, 1708, 1981, 1776]]

#
list = df.loc[(df['Year'] == 2019)]
list = list[21:]
print(list)
li2 = list.sort_values(by='Total', ascending=False)
li2.head(5)
sns.factorplot(x='Major area, region, country or area of destination', y='Total', data=li2, kind='bar', size=40,
               aspect=6)


# # 4-->Dynamics by income and Geographical Region

# # 1.Based on Geographical Region
#From following countries people migrated to these geographical regions
# running migration function for geographical region Africa,Asia,Europe,etc as a destination
demography = ["Africa", "Asia", "Europe", "Latin America and the Caribbean", "Northern America", "Oceania"]
years = [1990, 1995, 2000, 2005, 2010, 2015, 2019]
for i in range(len(demography)):
    for j in range(len(years)):
        migration(demography[i], years[j])

# # 2. Based on World Bank income groups
# running migration function for income criteria and High income,middle income and low income countries as destiantion

income = ["High-income countries", "Middle-income countries", "Low-income countries"]
years = [1990, 1995, 2000, 2005, 2010, 2015, 2019]
for i in range(len(income)):
    for j in range(len(years)):
        migration(income[i], years[j])



# # 2 and 3-->people from the more developed regions and less deveopled region are migrating to?

##function to find where people from less and more deveoped regions are migrating to
def more_developed(country, year):
    #printing country name and year
    print("Country:", country, "\tYear", year)
    cnt = []
    cnt.append(country)
    # column_names=["Year","Major area, region, country or area of destination","Belarus","Bulgaria","Czechia","Hungary","Poland","Republic of Moldova","Romania","Russian Federation","Slovakia","Ukraine","Channel Islands","Denmark","Estonia","Faroe Islands","Finland","Iceland","Ireland","Isle of Man","Latvia","Lithuania","Norway","Sweden","United Kingdom","Albania","Andorra","Bosnia and Herzegovina","Croatia","Gibraltar","Greece","Holy See","Italy","Malta","Montenegro","North Macedonia","Portugal","San Marino","Serbia","Slovenia","Spain","Austria","Belgium","France","Germany","Liechtenstein","Luxembourg","Monaco","Netherlands","Switzerland",,"New Zealand","Japan","Australia","Bermuda","Canada","Greenland","Saint Pierre and Miquelon","United States of America"]
    #selecting only given columns from dataframe following list is created
    column_names = ["Year", "Major area, region, country or area of destination"] + cnt
    if year == 1990:
        df2 = df[21:][column_names]
    elif year == 1995:
        df2 = df[304:][column_names]
    elif year == 2000:
        df2 = df[587:][column_names]
    elif year == 2005:
        df2 = df[870:][column_names]
    elif year == 2010:
        df2 = df[1153:][column_names]
    elif year == 2015:
        df2 = df[1435:][column_names]
    elif year == 2019:
        df2 = df[1719:][column_names]
    df2 = df2.loc[df["Year"] == year]
    #taking only country column and year values and sort them
    df2 = df2.sort_values(by=country, ascending=False)
    #printing top 3 countries people migrating from more/less developed regions
    df2.head(3)
    print(df2)

#followimg are the most developed countries in cntry list
cntry = ["Belarus", "Bulgaria", "Czechia", "Hungary", "Poland", "Republic of Moldova", "Romania", "Russian Federation",
         "Slovakia", "Ukraine", "Channel Islands", "Denmark", "Estonia", "Faroe Islands", "Finland", "Iceland",
         "Ireland", "Isle of Man", "Latvia", "Lithuania", "Norway", "Sweden", "United Kingdom", "Albania", "Andorra",
         "Bosnia and Herzegovina", "Croatia", "Gibraltar", "Greece", "Holy See", "Italy", "Malta", "Montenegro",
         "North Macedonia", "Portugal", "San Marino", "Serbia", "Slovenia", "Spain", "Austria", "Belgium", "France",
         "Germany", "Liechtenstein", "Luxembourg", "Monaco", "Netherlands", "Switzerland", "New Zealand", "Japan",
         "Australia", "Bermuda", "Canada", "Greenland", "Saint Pierre and Miquelon", "United States of America"]
years = [1990, 1995, 2000, 2005, 2010, 2015, 2019]
#running function to find where more developed regions people migrating to
for i in range(len(cntry)):
    for j in range(len(years)):
        more_developed(cntry[i], years[j])

# # people from the less developed regions are migrating to?
#followimg are the less developed countries in cntry list
cntry = ["Anguilla", "Antigua and Barbuda", "Aruba", "Bahamas", "Barbados", "British Virgin Islands",
         "Bonaire, Sint Eustatius and Saba", "Cayman Islands", "Cuba", "Curaçao", "Dominica", "Dominican Republic",
         "Grenada", "Guadeloupe", "Haiti", "Jamaica", "Martinique", "Montserrat", "Puerto Rico",
         "Saint Kitts and Nevis", "Saint Lucia", "Saint Vincent and the Grenadines", "Sint Maarten (Dutch part)",
         "Trinidad and Tobago", "Turks and Caicos Islands", "United States Virgin Islands"]
years = [1990, 1995, 2000, 2005, 2010, 2015, 2019]
#running function to find where less developed regions people migrating to
for i in range(len(cntry)):
    for j in range(len(years)):
        more_developed(cntry[i], years[j])

##--> Drilling down one of the region of my choice
# # CENTRAL AND SOUTHERN ASIA
#following are the countries included in Central and Southern Asia
cntry = ["Kazakhstan", "Kyrgyzstan", "Tajikistan", "Turkmenistan", "Uzbekistan", "Afghanistan", "Bangladesh", "Bhutan",
         "India", "Iran (Islamic Republic of)", "Maldives", "Nepal", "Pakistan", "Sri Lanka"]
years = [1990, 1995, 2000, 2005, 2010, 2015, 2019]

#to which countries people from Central and Southern Asia region migrating to
for i in range(len(cntry)):
    for j in range(len(years)):
        more_developed(cntry[i], years[j])

# #  maximum number of people are migrating from this countries
# #from which countries people are migrating to Central and Southern Asia region
complete = ["Kazakhstan", "Kyrgyzstan", "Tajikistan", "Turkmenistan", "Uzbekistan", "Afghanistan", "Bangladesh",
            "Bhutan", "India", "Iran (Islamic Republic of)", "Maldives", "Nepal", "Pakistan", "Sri Lanka"]
years = [1990, 1995, 2000, 2005, 2010, 2015, 2019]
for i in range(len(complete)):
    for j in range(len(years)):
        migration(complete[i], years[j])

# Plotting lineplot to find trend of people migrating from Central and Southern Asia
complete = ["Kazakhstan", "Kyrgyzstan", "Tajikistan", "Turkmenistan", "Uzbekistan", "Afghanistan", "Bangladesh",
            "Bhutan", "India", "Iran (Islamic Republic of)", "Maldives", "Nepal", "Pakistan", "Sri Lanka"]
for i in range(len(complete)):
    list = df.loc[df['Major area, region, country or area of destination'] == complete[i]]
    print(list)
    sns.lineplot(x='Year', y='Total', data=list)
    plt.show()

```
